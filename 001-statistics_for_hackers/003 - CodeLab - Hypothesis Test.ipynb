# Hypothesis Testing

Borrowing the case from previous codelab, about CTP and coin toss, we could rephrase our method to answer into a common statistical problem of Hypothesis Testing. 

Hypothesis testing, according to wolfram $^{[1]}$ is:
> the use of statistics to determine the probability that a given hypothesis is true

To do this usually we:
1. Determine the *null hypothesis* ($H_{0}$) and the *alternative hypothesis* ($H_{1})$.
2. Identify test statistics to assess the truth of the null hypothesis.
3. Compute $p$-value, which is the probability that a test statistics at least as significant as the one observed would be obtained.
4. Compare $p$-value to an $\alpha$. If $p \leq \alpha$ then null hypothesis will be accepted, else if $p > \alpha$ then null hypothesis will be rejected and alternative hypothesis will be accepted.

$H_0$ is usually set from our *a priori* knowledge, or **from our past knowledge**. So, in the coin problem, our $H_0$:
> The coin is a fair coin

We do know / assume that all coins are fair, because it always consists of 1 head and 1 tail.

$H_1$, on the other hand, means that the observations show a real effect combined with a component of chance variation. In the coin problem our $H_1$:
> The coin is not a fair coin




$H_0 = $ The coin is a fair coin

$H_1 = $ The coin is not a fair coin, more favor to head

We can re-write a more clearer definition of fair coin, from our observation. Let's take an example.

We know that a fair coin, with 30 tosses should result in about 15 heads. What is the probability of a fair coin showing 22 heads simply by chance?

To answer this, we need to reframe our assumptions as follows:
- A fair coin will favor less than 22 heads $P[X < 22]$.
- On the other hand, a non fair coin that favors head will translate to more than or equal to 22 heads $P[X \geq 22]$. 

If we set our confidence level to 95%^, which means if the $p$-value is lower than $0.05$, then **reject the null hypothesis** and we accept our alternative hypothesis that the coin is not fair.

^*just because there are many people using this confidence level, you can set it to 99% though*

Now we have set our **Confidence level** to 95%. What does it mean? 
1. We allow 5% Type I error $^{[3]}$. Type I error is **False Positive**, we **falsely reject null hypothesis** although it is **actually true**
2. We have set **Significant Level** $\alpha = 0.05$ or 5%


### Significant Level vs Confidence Level
The relationship between confidence level and significant level is
$$\text{Confidence Level} = 1 - \text{Significant Level}$$

Continuing to our discussion about coin, we want to calculate the $p$-value.

```python
# alpha is confidence level
alpha = 0.05
p_val = 1 - st.binom.cdf(n_head, n_tosses, p=prior)
msg  = 'p-value {:.3f} is '.format(p_val)

if p_val < alpha:
    msg += 'less than alpha {:.2f}, thus we reject the null hypothesis'.format(alpha)
else:
    msg += 'more than equal alpha {:.2f}, thus we accept the null hypothesis'.format(alpha)
    
print(msg)
```

This means we have reject null hypothesis and accept the alternative hypothesis.

According to this number, because the probability of getting 22 is super small, while we allow 5% False Positive, we have refused to believe that this coin is fair!
